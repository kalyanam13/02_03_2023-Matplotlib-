### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib. 
### Answer : Matplotlib:
### Matplotlib is a 2D plotting library for the Python programming language. It enables the creation of a wide variety of static, animated, and interactive visualizations in Python. Matplotlib is widely used for data visualization due to its flexibility and ease of use. It provides a wide range of plotting options and customization features.

### Uses of Matplotlib:

#### Visualization of data and trends.
#### Exploration and analysis of data.
#### Communication of findings through charts and graphs.
#### Five plots that can be created using the Pyplot module of Matplotlib:

### 1. Line Plot:

### Used for visualizing the relationship between two continuous variables over a continuous interval.
### Example: Plotting a time series.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.plot(x, y)
plt.show()
```

### 2. Scatter Plot:

### Used to display individual data points on a two-dimensional graph.
### Example: Visualizing the correlation between two variables.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.scatter(x, y)
plt.show()
```

### 3. Bar Plot:

### Used to represent categorical data with rectangular bars.
### Example: Comparing the performance of different categories.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values)
plt.show()
```

### 4. Histogram:

### Used to represent the distribution of a continuous variable.
### Example: Analyzing the distribution of exam scores.

```python
import matplotlib.pyplot as plt

data = [65, 70, 72, 78, 82, 85, 88, 90, 95]

plt.hist(data, bins=10)
plt.show()
```

### 5. Pie Chart:

### Used to represent the proportion of different categories in a whole.
### Example: Showing the percentage distribution of expenses.

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

### These are just a few examples, and Matplotlib provides a wide range of plot types and customization options to suit various visualization needs.






### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

### Note: Also add title, xlabel, and ylabel to the plot.
### Answer : Scatter Plot Explanation:
#### A scatter plot is a type of plot that displays individual data points on a two-dimensional graph. Each point represents the values of two variables, with one plotted along the x-axis and the other along the y-axis. Scatter plots are useful for visualizing the relationship or correlation between two continuous variables.

### Generating Data and Plotting Scatter Plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y, c='blue', marker='o', label='Data Points')

# Adding title, xlabel, and ylabel
plt.title('Scatter Plot of Randomly Generated Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the plot
plt.legend()
plt.show()
```

### Explanation of the code:

### 1. plt.scatter(x, y, c='blue', marker='o', label='Data Points'): This line creates a scatter plot using the generated data. The c parameter sets the color, marker specifies the marker style, and label provides a label for the data points.

### 2. plt.title('Scatter Plot of Randomly Generated Data'): Sets the title of the plot.

### 3. plt.xlabel('X-axis Label') and plt.ylabel('Y-axis Label'): Sets labels for the x-axis and y-axis, respectively.

### 4. plt.legend(): Displays the legend.

### 5. plt.show(): Shows the plot.

### You can run this code to generate and visualize the scatter plot with the provided data. Adjust the title, xlabel, and ylabel as needed for your specific application.

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
### Answer : subplot() Function:
### The subplot() function in Matplotlib is used to create multiple plots in the same figure. It helps in arranging and organizing multiple subplots within a single figure, allowing for better visualization and comparison of different datasets.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='orange')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='green')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='red')
plt.title('Line Plot 4')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

### Explanation:

### plt.subplot(2, 2, 1): Creates a 2x2 grid of subplots and selects the first subplot for Line Plot 1.
### plt.subplot(2, 2, 2): Selects the second subplot for Line Plot 2, and so on.
### plt.tight_layout(): Adjusts the layout to prevent overlapping of subplots.
### plt.show(): Displays the subplots.
### This code will generate a figure with four line plots arranged in a 2x2 grid, each representing different datasets. Adjust the labels, colors, and other plot parameters as needed.

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
### Answer : Bar Plot:
### A bar plot (or bar chart) is a graphical representation of categorical data with rectangular bars. Each bar's length or height corresponds to the value it represents, and the bars are typically separated by a small gap. Bar plots are used to compare the values of different categories, show the distribution of data, or display the relationship between categorical variables.

### Creating Bar Plot and Horizontal Bar Plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.title('Profit by Company (Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.title('Profit by Company (Horizontal Bar Plot)')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.show()
```

### Explanation:

### plt.bar(company, profit, color=['blue', 'green', 'red', 'purple']): Creates a vertical bar plot with the company names on the x-axis and their respective profits on the y-axis. The color parameter is used to set different colors for each bar.

### plt.barh(company, profit, color=['blue', 'green', 'red', 'purple']): Creates a horizontal bar plot with the company names on the y-axis and their respective profits on the x-axis.

### plt.title, plt.xlabel, plt.ylabel: Adds a title and labels to the plot for better understanding.

### plt.show(): Displays the plots.

### Adjust the figure size, colors, and other parameters as needed to suit your preferences. The bar plot and horizontal bar plot visually represent the profits of different companies, making it easy to compare the values.

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
### Answer : Box Plot:
### A box plot (or box-and-whisker plot) is a statistical visualization that provides a summary of the distribution of a dataset. It displays the median, quartiles, and potential outliers of the data. The plot consists of a box that represents the interquartile range (IQR) and "whiskers" that extend from the box to indicate the range of the data. Outliers are sometimes shown as individual points beyond the whiskers.
### Creating a Box Plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a single list for plotting
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Box')
plt.ylabel('Values')

# Show the plot
plt.show()
```

### Explanation:

### plt.boxplot(data, labels=['Box 1', 'Box 2']): Creates a box plot with two boxes (Box 1 and Box 2). The data list contains the values for each box, and the labels parameter sets the labels for each box.

### plt.title, plt.xlabel, plt.ylabel: Adds a title and labels to the plot for clarity.

### plt.show(): Displays the box plot.

### In this example, the box plot helps visualize the distribution of values in Box 1 and Box 2, making it easy to compare their central tendency, spread, and the presence of outliers. Adjust the figure size and other parameters as needed.
